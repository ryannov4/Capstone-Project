```python
# Daftar library
from langchain_community.llms import Replicate
import os
from google.colab import userdata
```

```python
# sample untuk granite
sample_df = df.sample(3).reset_index(drop=True)
```

```python
# Set the API token
api_token = userdata.get('REPLICATE_API_TOKEN')
os.environ["REPLICATE_API_TOKEN"] = api_token
# Model setup
model = "ibm-granite/granite-3.3-8b-instruct"
output = Replicate(
    model=model,
    replicate_api_token=api_token,
)
```

```python
summaries = []

for idx, row in sample_df.iterrows():
    article = row['Content']

    prompt = f"""
Summarize this financial news article:
a
{article}
"""
    response = output.invoke(prompt)

    # prefix dan baris baru biar rapi
    formatted_response = f"Summary {idx+1}:\n{response.strip()}\n"
    summaries.append(formatted_response)

sample_df['Granite_Summary'] = summaries
```

```python
summaries = []

for idx, row in sample_df.iterrows():
    article = row['Content']

    prompt = f"""
You are a financial analyst assistant.

Summarize the following financial news article using a structured format with clear section headings.

Include:

1. Main Events
2. Economic Indicators
3. Policy Implications
4. Market Impact
5. Recommendations (for investors or policymakers)

Be concise and formal. Use bullet points if needed. Do not copy the original text.

Article:

{article}
"""

    response = output.invoke(prompt)

    # prefix dan baris baru biar rapi
    formatted_response = f"Summary {idx+1}:\n{response.strip()}\n"
    summaries.append(formatted_response)

sample_df['Granite_Summary'] = summaries
```

```python
parameters = {
    "top_k": 5,                   # lebih fokus, tidak terlalu random
    "top_p": 0.9,                 # masih ada kreativitas tapi tidak terlalu liar
    "max_tokens": 220,            # batas panjang ringkasan
    "repetition_penalty": 1.2,    # kurangi pengulangan
    "min_tokens": 50,             # hindari ringkasan terlalu pendek
    "random_seed": 42             # hasil stabil kalau ulang
}
```

```python
summaries = []

for idx, row in sample_df.iterrows():
    article = row['Content'][:2000]

    # Add initial prompt
    refined_prompt = f"""
You are a financial analyst helping investors understand macroeconomic data.

Given this financial news article, do the following:
1. Identify and explain the key economic insights.
2. Highlight what these findings mean for investors.
3. Provide **clear and actionable investment recommendations** based on the analysis.

Present your output in a structured and concise format.

News Article:
{article}
"""


    # Invoke the model
    response = output.invoke(refined_prompt, parameters=parameters)

    # prefix dan baris baru biar rapi
    formatted_response = f"Summary {idx+1}:\n{response.strip()}\n"
    summaries.append(formatted_response)

sample_df['Granite_Refined_Summary'] = summaries
```
